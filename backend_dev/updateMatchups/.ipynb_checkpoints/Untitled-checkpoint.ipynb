```python
import requests
from bs4 import BeautifulSoup
import mysql.connector
import pandas as pd
from getLegend import getDay, getTeam
from theBot import theBot
import warnings
warnings.filterwarnings('ignore')



class getPredictions:
    def __init__(self, week, year):
        self.week = week
        self.year = year
        
        
        
    def teamService(self, team, upcomingWeekData):
        mydb = mysql.connector.connect(
            host="127.0.0.1",
            user="davidcarney",
            password="Sinorrabb1t",
            database="NFL"
            )

        mycursor = mydb.cursor()

        mycursor.execute(f"SELECT * FROM NFLRegularSeasons WHERE Team = '{team}'")

        result = mycursor.fetchall()
        
        dfColumns = ['ID','Week','Day','WinLoss','OT','At','OppTeam','Tm','Opp','Cmp','AttPassing','YdsPassing','TDPassing','Interceptions','Sk','YdsLossFromSacks','YPerAPassing','NYPerA','CmpPerc','Rate','AttRushing','YdsRushing','YPerARushing','TDRushing','FGM','FGA','XPM','XPA','Pnt','YdsPunting','ThirdDConv','ThirdDAtt','FourthDConv','FourthDAtt','ToP','Year','Team','YardsPerPoint','HomeTeam','Wind','Temp','Sky','Channel','GameLines','NeutralVenue','TotalScore']
        upcomingWeekColumns = ['Week','Day','At','OppTeam','Year']
        
        df = pd.DataFrame(result, columns = dfColumns)

        model = theBot(df, upcomingWeekColumns, [upcomingWeekData[0],upcomingWeekData[1],upcomingWeekData[2],upcomingWeekData[3],upcomingWeekData[4]])
        model.adjustDF()


        predictions1 = model.useCode()
        firstScore = list(predictions1.keys())[0]
        firstPredictions = predictions1[firstScore]

        predictions2 = model.useCode()
        secondScore = list(predictions2.keys())[0]
        secondPredictions = predictions2[secondScore]

        predictions3 = model.useCode()
        thirdScore = list(predictions3.keys())[0]
        thirdPredictions = predictions3[thirdScore]


        if (firstScore > secondScore) and (firstScore > thirdScore):
            if secondScore > thirdScore:
                return [[firstScore,firstPredictions], [secondScore,secondPredictions], [thirdScore,thirdPredictions]]
            else:
                return [[firstScore,firstPredictions], [thirdScore,thirdPredictions], [secondScore,secondPredictions]]

        elif (secondScore > firstScore) and (secondScore > thirdScore):
            if firstScore > thirdScore:
                return [[secondScore,secondPredictions], [firstScore,firstPredictions], [thirdScore,thirdPredictions]]
            else:
                return [[secondScore,secondPredictions], [thirdScore,thirdPredictions], [firstScore,firstPredictions]]

        else:
            if firstScore > secondScore:
                return [[thirdScore,thirdPredictions], [firstScore,firstPredictions], [secondScore,secondPredictions]]
            else:
                return [[thirdScore,thirdPredictions], [secondScore,secondPredictions], [firstScore,firstPredictions]]

        
        
        
    def updateMatchup(self, matchup, resultsFile):
        allTeams = {'Cowboys':'dal','Buccaneers':'tam','Saints':'nor','Rams':'ram','Lions':'det','Falcons':'atl','Panthers':'car','Vikings':'min','Packers':'gnb','Bears':'chi','Cardinals':'crd','49ers':'sfo','Seahawks':'sea','Titans':'oti','Colts':'clt','Jaguars':'jax','Chiefs':'kan','Raiders':'rai','Chargers':'sdg','Broncos':'den','Eagles':'phi','Washington':'was','Giants':'nyg','Jets':'nyj','Dolphins':'mia','Patriots':'nwe','Bills':'buf','Bengals':'cin','Steelers':'pit','Ravens':'rav','Browns':'cle','Texans':'htx'}

        homeTeam = ''
        awayTeam = ''

        homeTeamName, awayTeamName, day = matchup

        for team in allTeams:
            if team in homeTeamName:
                homeTeam = team

            if team in awayTeamName:
                awayTeam = team


        overallHomeScore = 0
        overallAwayScore = 0

        for i in range(12):

            homeTeamList = self.teamService(allTeams[homeTeam], [self.week,getDay(day),1,getTeam(awayTeamName),self.year])
            awayTeamList = self.teamService(allTeams[awayTeam], [self.week,getDay(day),0,getTeam(homeTeamName),self.year])

            for homePrediction, awayPrediction in zip(homeTeamList, awayTeamList):
                homeAvgClfReg = (homePrediction[1][0] + homePrediction[1][1]) / 2
                awayAvgClfReg = (awayPrediction[1][0] + awayPrediction[1][1]) / 2

                
                homeScore = homeAvgClfReg
                awayScore = awayAvgClfReg

                overallHomeScore = overallHomeScore + homeScore[0]
                overallAwayScore = overallAwayScore + awayScore[0]


        if overallHomeScore > overallAwayScore:
            percent = (1 - (overallAwayScore / overallHomeScore)) * 100
            #print(f"Percent: {percent}% for {homeTeam}")
            #return Response([homeTeam, awayTeam, round(percent,2)])
            resultsFile.write(awayTeam)
            resultsFile.write(',')
            resultsFile.write(homeTeam)
            resultsFile.write(',')
            resultsFile.write(homeTeam)
            resultsFile.write(',')
            resultsFile.write(str(round(percent,2)))
            resultsFile.write('\n')
        else:
            percent = (1 - (overallHomeScore / overallAwayScore)) * 100
            #print(f"Percent: {percent}% for {awayTeam}")
            #return Response([awayTeam, homeTeam, round(percent,2)])
            resultsFile.write(awayTeam)
            resultsFile.write(',')
            resultsFile.write(homeTeam)
            resultsFile.write(',')
            resultsFile.write(awayTeam)
            resultsFile.write(',')
            resultsFile.write(str(round(percent,2)))
            resultsFile.write('\n')
        
        
        
    def doit(self):
        matchups = []
        readFilename = "upcomingWeekData.txt"
        resultFilename = "results.txt"
        
        readFile = open(readFilename, "r")
        resultsFile = open(resultFilename, "a")

        for matchup in readFile.readlines():
            homeTeam, awayTeam, day = matchup.split("_")
            day = day.strip()

            matchups.append([homeTeam, awayTeam, day])

        for matchup in matchups:
            self.updateMatchup(matchup, resultsFile)
            
        readFile.close()
        resultsFile.close()
```

```python
week = 6
year = 2022
obj = getPredictions(week, year)
obj.doit()
```
